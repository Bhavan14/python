## **Now we use replace or LabelEncoding to change object to int**
    If the unique values less than 10 different values we use replace more then 10 difference values we use LabelEncoding
    We use LabelEncoding for [model,Location] balance we use replace 

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#array(['skoda', 'fiat', 'bmw', 'nissan', 'audi', 'toyota', 'hyundai','maserati'], dtype=object)
```

```python
df['Maker']=df['Maker'].replace({'skoda':0, 'fiat':1, 'bmw':2, 'nissan':3, 'audi':4, 'toyota':5, 'hyundai':6,'maserati':7})
```
